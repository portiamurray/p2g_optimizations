```python
DGSHP=pd.DataFrame(index=range(1,4),columns=['GPotential'])
TotalRoofArea=pd.DataFrame(index=range(1,4),columns=['RoofArea'])
totalPV=pd.DataFrame(index=range(0,8760),columns=range(1,4))
totalST=pd.DataFrame(index=range(0,8760),columns=range(1,4))
for i in range(1,4):
    group='Group'+str(i)
    snodes=pd.read_excel(path,group,parse_cols='A:E')
    Buildings=snodes[snodes['Type']=='Building']
    DGSHP.loc[i,'GPotential']=GSHP.loc[Buildings.index,'Gpotential'].sum()
    TotalRoofArea.loc[i,'RoofArea']=RoofArea.loc[Buildings.index,'Area'].sum()
    tPV=PV.loc[:,Buildings.index]*RoofArea.loc[Buildings.index,'Area']
    tST=ST.loc[:,Buildings.index]*RoofArea.loc[Buildings.index,'Area']
    if TotalRoofArea.loc[i,'RoofArea']>0:
        totalPV.loc[:,i]=tPV.sum(axis=1)/TotalRoofArea.loc[i,'RoofArea']
        totalST.loc[:,i]=tST.sum(axis=1)/TotalRoofArea.loc[i,'RoofArea']
    else:
        totalPV.loc[:,i]=0
        totalST.loc[:,i]=0
DGSHP
```
